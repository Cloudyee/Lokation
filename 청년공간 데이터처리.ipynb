```python
## 필요한 라이브러리 임포트
import requests
import pprint
```

```python
url = "https://www.youthcenter.go.kr/opi/wantedSpace.do?openApiVlak=3ea4606804415b8f4505d137&pageNo=1&numOfRows=1000&type=xml"
```

```python
response = requests.get(url)
contents = response.text
```

```python
## 보다 예쁘게 출력하기 위해..
##  indent는 들여쓴다는 의미
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
### xml을 DataFrame으로 변환하기 ###
from os import name
import xml.etree.ElementTree as et
import pandas as pd
import bs4
from lxml import html
from urllib.parse import urlencode, quote_plus, unquote
```

```python
## 각 컬럼 값 ## (포털 문서)
"""
totalCnt : 총건수
pageIndex : 조회된 현재 페이지
rownum : 해당 데이터의 번호
spcId : 공간ID
spcName : 공간명
areaCpvn : 지역 - 시, 도
areaSggn : 지역 - 시,군,구
address : 주소
spcTime : 공간이용시간
operOrgan : 운영기관
homepage : 홈페이지 URL
telNo : 연락처
officeHours : 업무시간
openDate : 개소일
applyTarget : 지원대상
spcType : 공간유형
majorForm : 주요형태
spcCost : 비용
addFacilCost : 부대시설 비용(유료,무료)
foodYn : 식음료 제공 여부
""" 

#bs4(Beautiful Soup) 사용하여 item 태그 분리

xml_obj = bs4.BeautifulSoup(contents,'lxml-xml')
rows = xml_obj.findAll('spacesInfo')
print(rows)
```

```python
# 데이터를 담을 빈 리스트 생성
data = []

# 각 'spacesInfo' 태그를 순회하며 데이터 추출
for row in rows:
    # 각 컬럼의 값을 추출
    totalCnt = row.find('totalCnt').text
    pageIndex = row.find('pageIndex').text
    rownum = row.find('rownum').text
    spcId = row.find('spcId').text
    spcName = row.find('spcName').text
    areaCpvn = row.find('areaCpvn').text
    areaSggn = row.find('areaSggn').text
    address = row.find('address').text
    spcTime = row.find('spcTime').text
    operOrgan = row.find('operOrgan').text
    homepage = row.find('homepage').text
    telNo = row.find('telNo').text
    officeHours = row.find('officeHours').text
    openDate = row.find('openDate').text
    applyTarget = row.find('applyTarget').text
    spcType = row.find('spcType').text
    majorForm = row.find('majorForm').text
    spcCost = row.find('spcCost').text
    addFacilCost = row.find('addFacilCost').text
    foodYn = row.find('foodYn').text

```

```python
# 각 컬럼의 값을 포함하는 데이터 리스트 생성
row_data = [totalCnt, pageIndex, rownum, spcId, spcName, areaCpvn, areaSggn, address, spcTime, operOrgan,
                homepage, telNo, officeHours, openDate, applyTarget, spcType, majorForm, spcCost, addFacilCost, foodYn]
# 데이터 리스트에 추가
data.append(row_data)
```

```python
# 데이터프레임 생성
df = pd.DataFrame(data, columns=['totalCnt', 'pageIndex', 'rownum', 'spcId', 'spcName', 'areaCpvn', 'areaSggn', 'address', 'spcTime', 'operOrgan',
                                 'homepage', 'telNo', 'officeHours', 'openDate', 'applyTarget', 'spcType', 'majorForm', 'spcCost', 'addFacilCost', 'foodYn'])
```

```python
# 데이터프레임 출력
print(df)

```
