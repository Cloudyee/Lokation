```python



# 데이터프레임 주소값 추출
address= local_workation['address']
print(address.head(10))


```

```python
!pip install geopy

####### 도로명주소 위도 경도 값으로 바꿔주기 ########
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
#####주소를 위,경도 값으로 변환하기 #####
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
print(geo_local)
```

```python
#####Dataframe만들기######
geocoding_workation = pd.DataFrame({'name': local_workation['name'],'option':local_workation['option'],'address':address,'latitude':latitude,'longitude':longitude})

#df저장
geocoding_workation.to_csv('local_workation_geocoding.csv')
```
