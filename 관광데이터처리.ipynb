```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 각 카테고리의 개수 계산
category_counts = category['category'].value_counts()
```

```python
category_counts
```

```python
# 시각화 설정
plt.figure(figsize=(8, 6))
plt.bar(category_counts.index, category_counts.values, color='skyblue')
plt.xlabel('카테고리')
plt.ylabel('개수')
plt.title('카테고리별 개수')

# 그래프 표시
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 'category'는 사용자의 데이터프레임, 'category'는 사용자의 컬럼명입니다.
sns.set(style="darkgrid")
ax = sns.countplot(x="category", data=category, palette="Set2")
plt.title('카테고리별 개수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 사용할 글꼴 설정
plt.rcParams['font.family'] = 'NanumGothic'

# 데이터 및 색상 설정
colors = ['skyblue', 'salmon', 'lightgreen']  # 각 카테고리에 대한 색상 설정

# 그래프 생성 및 설정
plt.figure(figsize=(8, 6))
bars = plt.bar(category_counts.index, category_counts.values, color=colors)
plt.ylabel('개수')
plt.title('카테고리별 개수')

# 각 막대에 레이블 추가
for bar, count in zip(bars, category_counts.values):
    plt.text(bar.get_x() + bar.get_width() / 2, count, str(count), ha='center', va='bottom')

# 그래프 표시
plt.show()
```
